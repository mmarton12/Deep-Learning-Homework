## Exercise 4.

First we have to include the necessary libraries. Then using the URL of the Deep Learning wikipedia page we send a request, parse the html and extract the text.

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Deep_learning"      #sets the url to the Deep Learning wikipedia page
html = urlopen(url).read()                               #reads the response object for the url requested
soup = BeautifulSoup(html, features="html.parser")       #creates a BeautifulSoup object from html using Python's html.parser

text = soup.get_text()                                   #extracts the text

print(text)
```

To plot the letter frequency on a histogram first we need to filter out only the english letters then we need to measure their frequency and then we can plot it.

```python
sfiltered = ''

for char in text.upper():
    if((ord(char) >= 97 and ord(char) <= 122) or (ord(char) >= 65 and ord(char) <= 90)):     #makes all letters upper case and filters out the english ones
        sfiltered += char

count = Counter(sfiltered)                                              #creates a dictionary containing each number and its frequency
count_sorted = {key:value for key, value in sorted(count.items())}      #sorts the dictionary alphabetically
plt.bar(count_sorted.keys(), count_sorted.values(), color='g');         #plots the letter frequency on a histogram
```
